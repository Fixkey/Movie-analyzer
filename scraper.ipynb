```python
import pandas as pd

ratings = pd.read_csv("results/ratings.csv")

ratings.describe()
```

```python

import scrapy
from scrapy.crawler import CrawlerProcess

class MySpider(scrapy.Spider):
    name = 'mySpider'
    start_urls = [f'https://www.imdb.com/title/{id}/' for id in ratings['id']]

    custom_settings = {
        'LOG_LEVEL': 'INFO',
        # 'ITEM_PIPELINES': {'__main__.JsonWriterPipeline': 2},
        'REQUEST_FINGERPRINTER_IMPLEMENTATION': '2.7',
        'FEEDS': {'./results/plots.json': {'format': 'json', 'overwrite': True}} 
    }
    
    def parse(self, response):
        item = {
            'plot':  'N/A',
            'id': response.url.split('/')[4]
        }
        
        print(f"Response status: {response.status}")
        item['plot'] = response.css('span[data-testid="plot-xl"]::text').get(default='N/A')
        
        yield item


process = CrawlerProcess({
    'USER_AGENT':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36',
    'REQUEST_FINGERPRINTER_IMPLEMENTATION': '2.7',
})

process.crawl(MySpider)
process.start()
```
